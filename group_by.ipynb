```python
diccionario_mapa = {0: "No", 1: "Si"} 
data3["uses_ad_boosts"] = data3["uses_ad_boosts"].map(diccionario_mapa)
data3["badge_local_product"] = data3["badge_local_product"].map(diccionario_mapa)
data3["badge_product_quality"] = data3["badge_product_quality"].map(diccionario_mapa)
data3["badge_fast_shipping"] = data3["badge_fast_shipping"].map(diccionario_mapa)
data3["shipping_is_express"] = data3["shipping_is_express"].map(diccionario_mapa)
```

```python
# Calificación de la calidad del producto. Sobreescribe la columna badge_product_quality utilizando apply() para asignar "Buena" a las filas donde 'badge_product_quality' sea "No" y "Excelente" donde sea "Si".
```

```python
badge_product_quality = data3['badge_product_quality'].apply(lambda x: 'Buena' if x == 'No' else 'Excelente')
data3['badge_product_quality'] = badge_product_quality
```

```python
# Cálculo de descuento. Crea una nueva columna discount_percentage utilizando apply() que calcule el porcentaje de descuento basado en las columnas 'price' y 'retail_price', redondea los decimales a 2. Después elimina las columnas de average_discount y mean_discount
```

```python
def calcular_desc(fila):

    resultado = fila['price']*fila['retail_price']*100
    resultado = round(resultado, 2)
    return resultado
```

```python
data3["discount_percentage"] = data3.apply(calcular_desc, axis = 1)
data3.head()
```

```python
# Evaluación de la valoración del vendedor. Utiliza apply() para crear una nueva columna 'seller_reputation' donde los valores sean "Buena" si 'merchant_rating' es mayor o igual a 4, y "Regular" si es menor a 4. Después elimina la columna merchant_rating.
```

```python
def valoracion_vendedor(merchant_rating):
    if merchant_rating >= 4:
        return "Buena"
    else:                                       
        return "Regular"

    
```

```python
data3["seller_reputacion"] = data3["merchant_rating"].apply(valoracion_vendedor)    
data3.drop(columns=['merchant_rating'], inplace=True)
data3.head()
```
